```python
import tensorflow as tf

tensor1 = tf.constant([[1, 2, 3],
                        [4, 5, 6]])

tensor2 = tf.constant([[7, 8, 9],
                        [10, 11, 12]])

tensor_sum = tf.add(tensor1, tensor2)

print(tensor_sum)
```

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [10, 11, 12]])

C = A + B

print(C)
```

```python
import numpy as np

# Create two (2,3) tensors with random values
A = np.random.rand(2, 3)
B = np.random.rand(2, 3)

# Element-wise addition
C = A + B

print("Tensor A:\n", A)
print("\nTensor B:\n", B)
print("\nElement-wise Addition (A + B):\n", C)
```
